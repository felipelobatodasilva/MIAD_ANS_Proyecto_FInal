```python
def get_release_date(song_name):
    # Buscar la canción en Spotify
    results = sp.search(q=song_name, type='track', limit=1)
    
    if results['tracks']['items']:
        track = results['tracks']['items'][0]
        release_date = track['album']['release_date']
        return release_date
    else:
        return 'Canción no encontrada'

def get_artist_followers(artist_name):
    #Obtener seguirdores de spotify
    results = sp.search(q=artist_name, type='artist', limit=1)
    if results['artists']['items']:
        artist = results['artists']['items'][0]
        followers = artist['followers']['total']
        return followers
    else:
        return 'Artista no encontrado'
  
# Crear listas vacías para almacenar la información
release_dates = []
followers_counts = []

# Obtener información adicional
release_dates = [get_release_date(song) for song in Cancion]
followers_counts = [get_artist_followers(artist) for artist in Artista]

Crear DataFrames para hacer el merge_ para traer fecha de lanzamiento y # de seguidores.
df_release_dates = pd.DataFrame({
    'cancion': Cancion,
    'fecha_lanzamiento': release_dates
})

df_followers = pd.DataFrame({
    'artista': Artista,
    'Followers': followers_counts
})

# Unir los DataFrames con el DataFrame original
df_combined = df_spotify.merge(df_release_dates, on='cancion', how='left')
df_combined = df_combined.merge(df_followers, on='artista', how='left')

#Creando el campo dia_semana co la fecha de lanzamiento
# Suponiendo que tu columna se llama 'fecha_lanzamiento'
df_combined['fecha_lanzamiento'] = pd.to_datetime(df_combined['fecha_lanzamiento'])

# Extraer el día de la semana
df_combined['dia_semana'] = df_combined['fecha_lanzamiento'].dt.day_name()
```

```python
def get_top_tracks_from_playlist(playlist_id):
    results = sp.playlist_tracks(playlist_id)
    top_tracks = [track['track']['artists'][0]['name'] for track in results['items']]
    return top_tracks


def check_if_artist_in_top(artist_name, artists_set):
    return 'sí' if artist_name in artists_set else 'no'

#IDs de las listas de reproducción del Top 50
playlist_ids_2019 = ['37i9dQZF1DXcBWIGoYBM5M']  # Top 50 en EE.UU., por ejemplo
playlist_ids_2020 = ['37i9dQZF1DXcBWIGoYBM5M']  # Top 50 en EE.UU., por ejemplo

# Obtener los artistas de las listas de reproducción
artists_2019 = set()
artists_2020 = set()

for playlist_id in playlist_ids_2019:
    artists_2019.update(get_top_tracks_from_playlist(playlist_id))

for playlist_id in playlist_ids_2020:
    artists_2020.update(get_top_tracks_from_playlist(playlist_id))

# Supongamos que df_spotify es tu DataFrame original
df_combined['TOP50_2019-2020'] = df_combined['artista'].apply(lambda x: check_if_artist_in_top(x, artists_2019.union(artists_2020)))
```

### Informacion MacroEconomica

```python
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Leer los archivos CSV
songs_df = pd.read_csv("df_combined.csv")
inflation_df = pd.read_csv("Inflation_USA_worldbank.csv")
unemployment_df = pd.read_csv("unemployment_worldbank.csv")
gdp_df = pd.read_csv("gdp-per-capita-worldbank.csv")

# Limpiar la columna de GDP per capita
gdp_df.columns = gdp_df.columns.str.strip()
gdp_df['GDP per capita'] = gdp_df['GDP per capita'].str.replace(',', '', regex=False)  # Asegurar el reemplazo de comas
gdp_df['GDP per capita'] = pd.to_numeric(gdp_df['GDP per capita'], errors='coerce')

# Renombrar columnas para la consistencia
inflation_df = inflation_df.rename(columns={'Year': 'anio'})
unemployment_df = unemployment_df.rename(columns={'Year': 'anio'})
gdp_df = gdp_df.rename(columns={'Year': 'anio'})

# Convertir la columna 'anio' a entero en todos los DataFrames
inflation_df['anio'] = inflation_df['anio'].astype(int)
unemployment_df['anio'] = unemployment_df['anio'].astype(int)
gdp_df['anio'] = gdp_df['anio'].astype(int)
songs_df['anio'] = songs_df['anio'].astype(int)


# Hacer merge con base en 'anio'
songs_with_inflation = pd.merge(songs_df, inflation_df[['anio', 'Inflation, consumer prices (annual %)']], 
                                on='anio', how='left')

songs_with_unemployment = pd.merge(songs_with_inflation, unemployment_df[['anio', 'Unemployment (%)']], 
                                   on='anio', how='left')

songs_complete = pd.merge(songs_with_unemployment, gdp_df[['anio', 'GDP per capita']], 
                          on='anio', how='left')

# Renombrar las columnas para mayor claridad
songs_complete_economics = songs_complete.rename(columns={
    'Inflation, consumer prices (annual %)': 'inflation',
    'Unemployment (%)': 'unemployment',
    'GDP per capita': 'gdp_per_capita'
})



# Verificar los resultados
print("Primeras filas del DataFrame resultante:")
print(songs_complete_economics.head())
songs_complete_economics
```
