```python
def get_release_date(song_name):
    # Buscar la canción en Spotify
    results = sp.search(q=song_name, type='track', limit=1)
    
    if results['tracks']['items']:
        track = results['tracks']['items'][0]
        release_date = track['album']['release_date']
        return release_date
    else:
        return 'Canción no encontrada'

def get_artist_followers(artist_name):
    # Obtener seguidores de Spotify
    results = sp.search(q=artist_name, type='artist', limit=1)
    if results['artists']['items']:
        artist = results['artists']['items'][0]
        followers = artist['followers']['total']
        return followers
    else:
        return 'Artista no encontrado'
  
# Crear listas vacías para almacenar la información
release_dates = []
followers_counts = []

# Obtener información adicional
release_dates = [get_release_date(song) for song in Cancion]
followers_counts = [get_artist_followers(artist) for artist in Artista]

# Crear DataFrames para hacer el merge y traer fecha de lanzamiento y seguidores
df_release_dates = pd.DataFrame({
    'cancion': Cancion,
    'fecha_lanzamiento': release_dates
})

df_followers = pd.DataFrame({
    'artista': Artista,
    'Followers': followers_counts
})

# Unir los DataFrames con el DataFrame original
df_combined = df_spotify.merge(df_release_dates, on='cancion', how='left')
df_combined = df_combined.merge(df_followers, on='artista', how='left')

# Ajustar los valores de la columna 'fecha_lanzamiento' para manejar diferentes formatos
# Convertir los años que están solos al formato de fecha
df_combined['fecha_lanzamiento'] = df_combined['fecha_lanzamiento'].apply(lambda x: f"{x}-01-01" if isinstance(x, str) and len(x) == 4 else x)

# Convertir a datetime, manejando diferentes formatos
df_combined['fecha_lanzamiento'] = pd.to_datetime(df_combined['fecha_lanzamiento'], errors='coerce', format='mixed')

# Extraer el día de la semana
df_combined['dia_semana'] = df_combined['fecha_lanzamiento'].dt.day_name()

# Mostrar el DataFrame resultante
print(df_combined)
```

## 3. Analisis de cluster

El objetivo principal de este análisis es perfilar el tipo de música entre los años 2000 y 2020 durante épocas de bonanza económica y exansión economica en EEUU. Para ello, primero se creará una variable que identifique los periodos de bonanza según la fecha de lanzamiento de las canciones. Esta información se obtendrá a partir de los datos proporcionados por el Comité de Datación del Ciclo Económico de la NBER, que publica una cronología oficial de los ciclos económicos en Estados Unidos. 

![cronología del NBER](cicloeeuu.png)

La cronología del NBER define los picos y valles que delimitan las fases de recesión y expansión, lo que nos permitirá clasificar los periodos de expansión como épocas de bonanza. Con esta variable, podremos analizar las características musicales predominantes en estos periodos prósperos.

```python
# Datos del año y recesión
data = {
    'año': [1999, 2000, 2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 
            2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020],
    'recesion': ['no', 'no', 'si', 'no', 'no', 'no', 'no', 'no', 'no', 'no', 'si', 
                 'no', 'no', 'no', 'no', 'no', 'no', 'no', 'no', 'no', 'no', 'si']
}
df_recesion = pd.DataFrame(data)

#Añadiendo la vble recesion a songs_complete_final
songs_complete_final['año'] = songs_complete_final['anio'].astype(int)
df_recesion['año'] = df_recesion['año'].astype(int)
df_merge = pd.merge(songs_complete_final, df_recesion[['año', 'recesion']], on='año', how='left')
#Eliminando año del df
df_merge = df_merge.drop('año', axis=1)
```

Ahora Dividiremos el df en años de recesión y años de no recesión para analizar cómo varía el perfil musical en función de las condiciones económicas. Este enfoque permitirá aplicar el clustering por separado a cada conjunto de datos y observar posibles diferencias en el comportamiento musical.

### 3.2 caracteristicas musicas epoca de recesión

#### 3.2.1 Reducción de dimensionalidad con PCA

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Suponiendo que X_scaled es tu matriz de datos estandarizados
pca = PCA()
pca.fit(df_recesion_si)

# Varianza explicada acumulativa
explained_variance_ratio = pca.explained_variance_ratio_.cumsum()

# Gráfico del codo
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(explained_variance_ratio) + 1), explained_variance_ratio, marker='o')
plt.title('Gráfico del Codo para Selección de Componentes Principales')
plt.xlabel('Número de Componentes Principales')
plt.ylabel('Varianza Explicada Acumulativa')
plt.grid(True)
plt.show()
```

 Dado que la varianza explicada alcanza un valor del 99% a partir de 15 componentes, indica que estos 15 componentes son suficientes para capturar la mayor parte de la información en los datos. Se Puede considerar usar estos 15 componentes para el análisis y reducir la dimensionalidad sin perder casi nada de la varianza original.

```python
# Ajusta PCA con el número de componentes que seleccionaste
n_components = 15
pca = PCA(n_components=n_components)
X_pca = pca.fit_transform(df_recesion_si)

# Convertir el resultado a un DataFrame para facilitar su uso posterior
pca_columns = [f'PC{i+1}' for i in range(n_components)]
df_pca = pd.DataFrame(X_pca, columns=pca_columns)        
```

#### 3.2.2 Realizar cluster jerarquico

Lo primero que haremos es hacer clustering jerárquico, esto ayudará a obtener una visión general de cómo se agrupan los datos sin necesidad de especificar el número de clusters de antemano. Esto puede proporcionar información útil sobre la estructura subyacente de los datos.

```python
#Calculando la Matriz de Distancias: Se utilizara la distancia Euclidiana, para calcular las distancias entre los puntos de datos.
import scipy.cluster.hierarchy as sch
import matplotlib.pyplot as plt
distance_matrix = sch.distance.pdist(X_pca)
```

```python
#Construyendo el Dendrograma:para visualizar las jerarquías y decidir el número adecuado de clusters.
linkage_matrix = sch.linkage(distance_matrix, method='ward')
plt.figure(figsize=(10, 6))
sch.dendrogram(linkage_matrix)
plt.title('Dendrograma para Clustering Jerárquico')
plt.xlabel('Índice de la Muestra')
plt.ylabel('Distancia')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering

#Cortar el Dendrograma: Decide el número de clusters cortando el dendrograma en el nivel adecuado.
n_clusters = 20# Aplicar clustering jerárquico con el número de clusters decidido
hierarchical = AgglomerativeClustering(n_clusters=n_clusters, linkage='ward')
hierarchical_clusters = hierarchical.fit_predict(X_pca)

# Añadir los clusters al DataFrame
df_pca['hierarchical_cluster'] = hierarchical_clusters
```
